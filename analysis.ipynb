### Working with `cast`

```python
df['cast']
```

```python
all_actors = []
for row in df['cast']:
    if isinstance(row, str):
        row = row.split(",")
        all_actors.extend(row)

for i, actor in enumerate(all_actors):
    all_actors[i] = actor.strip()

distinct_actors = list(set(all_actors))
distinct_actors
```
