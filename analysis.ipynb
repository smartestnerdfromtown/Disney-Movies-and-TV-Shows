#### Dropping Unnecessary Columns

```python
# If there are no duplicates, then, we can drop show_id. Besides this, there is not need for keeping show_id
df = df.drop(labels='show_id', axis='columns')

# I am not going to use somehow description column.
df = df.drop(labels='description', axis='columns')
```

#### Fixing Data Types

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df
```

### Working with `director`

```python
all_directors = []
for row in df['director']:
    if isinstance(row, str):
        row = row.split(",")
        all_directors.extend(row)

for i, director in enumerate(all_directors):
    all_directors[i] = director.strip()

distinct_directors = list(set(all_directors))
```

```python
print(f"Number of distinct directors: {len(distinct_directors)}")
```

```python
directors_and_count = {director: all_directors.count(director) for director in distinct_directors}
directors_and_count = dict(
    sorted(
        directors_and_count.items(), key=lambda item: item[1], reverse=True
))
```

### Working with `cast`

```python
df['cast']
```

```python
all_actors = []
for row in df['cast']:
    if isinstance(row, str):
        row = row.split(",")
        all_actors.extend(row)

for i, actor in enumerate(all_actors):
    all_actors[i] = actor.strip()

distinct_actors = list(set(all_actors))
```

```python
print(f"Number of actors: {len(distinct_actors)}")
```

```python
actors_and_count = {actor: all_actors.count(actor) for actor in distinct_actors}
actors_and_count = dict(
    sorted(
        actors_and_count.items(), key=lambda item: item[1], reverse=True
))
```

```python
def find_movies_by_actor(actor_name: str):
    for record in df[["title", "cast"]].itertuples(index=False):
        if actor_name in str(record.cast):
            print(record.title)
```

```python
find_movies_by_actor("Walt Disney")
```
